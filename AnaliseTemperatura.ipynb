```python
import numpy as np
```

```python
class MonitorDeTemperatura:
  """
    Classe para monitorar/adicionar temperaturas, calcular estatísticas e
    verificar se estão em um intervalo seguro de grau
  """

  def __init__(self):
    self.listaTemperaturas = []

  def AdicionarTemperaturas(self, valorTemp):
    self.listaTemperaturas.append(valorTemp)
    print("A nova temperatura foi adicionada com sucesso!")

  def CalcularEstatisticaTemp(self):
    dados = np.array(self.listaTemperaturas)

    analise_estatistica = {
        "Máxima" : dados.max(),
        "Mínima" : dados.min(),
        "Média" : dados.mean()
    }

    return analise_estatistica;

  def VerificarTemp(self):
    #condição para verificar se as temperaturas estão entre 20 e 80 graus

    foraLimite = False;
    for temp in self.listaTemperaturas:
      if temp < 20 or temp > 80:
        print(f"Alerta de temperatura detectado: {temp}°C ")
        foraLimite = True
    if not foraLimite:
      print("Todas as temperaturas estão dentro de um limite seguro")
```

```python
monitor = MonitorDeTemperatura()

print("Bem vindo ao programa de análise de temperaturas!")

valores_temp = int(input("Informe quantas temperaturas você deseja adicionar:"))

for temp in range(valores_temp):
  valorTempInformado = float(input(f"Informe a {temp + 1}ª temperatura: "))
  monitor.AdicionarTemperaturas(valorTempInformado)

estatistica = monitor.CalcularEstatisticaTemp();

print("\nEstatísticas das temperaturas registradas:")
print(f"Temperatura Máxima: {estatistica["Máxima"]}°C")
print(f"Temperatura Mínima: {estatistica["Mínima"]}°C")
print(f"Temperatura Média: {estatistica["Média"]}°C")

monitor.VerificarTemp()
```
